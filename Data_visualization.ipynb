# **Part Two: Data Visualization**:

# ● **Bar chart**
###  to show customer distribution

```python
print(df['Gender'].unique())
```

```python
print(df["Gender"].value_counts())
```

```python
plt.figure(figsize=(6, 6))
ax = sns.countplot(data=df, x="Gender", palette="coolwarm", alpha=0.7)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=12)

plt.title("Distribution of Customers by Gender")
plt.xlabel('Gender') 
plt.ylabel('Count')
plt.show()
```

# ● **Histogram chart**
###  to show the age distribution of customers

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='Age', bins=20, hue='Gender', kde=True,
            multiple="stack", alpha =0.7, palette='coolwarm')

plt.title('Age Distribution of Customers', fontsize=14)
plt.xlabel('Age')
plt.ylabel('count')

plt.show()
```

# ● **Boxplot chart**
###  to show the distribution of total shopping expenses among different genders

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Gender',y='Total_Spent', palette="coolwarm")

plt.title('Distribution of Total Spent by Gender', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

# ● **Bar chart**
###  to display average purchase cost by geographic region

```python
plt.figure(figsize=(10,8))
axR = sns.barplot(data=df, x='Region', y='Total_Spent', estimator=lambda x : x.mean(), errorbar=None,
            alpha=0.7, palette="coolwarm")

for i in axR.patches:
    axR.annotate(f'{i.get_height():2f}', (i.get_x() + i.get_width()/2, i.get_height()),
                 ha='center', va='bottom', fontsize=10)
                 
plt.title('Average Total by Region') 
plt.show()
```

# ● **Scatterplot**
###  to Analyzing the Relationship Between Purchase Frequency and Total Spending

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='Purchase_Frequency', y='Total_Spent', hue='Total_Spent',
                palette="magma_r")

plt.title('Analyzing the Relationship Between Purchase Frequency and Total Spending')
plt.xlabel('Purchase_Frequency')
plt.ylabel('Total_Spent')
plt.grid()
plt.show()
```

# ● **Countplot chart**
###  to display the most popular product categories

```python
plt.figure(figsize=(10,8))
axP = sns.countplot(data=df, x='Product_Category', palette='coolwarm', alpha=0.5)

for a in axP.patches:
    axP.annotate(f'{a.get_height()}', (a.get_x() + a.get_width()/2 , a.get_height())
                 , ha='center', va='bottom', fontsize= 10)
    
plt.title('Most popular products')
plt.xlabel('Product_Category')
plt.ylabel('Count')
plt.show()
```

# ● **Bar chart**
## to display average customer ratings in each product category

```python
plt.figure(figsize=(10,8))
axC = sns.barplot(data=df, x='Product_Category', y='Customer_Rating', estimator= lambda y : y.mean()
                  , palette='coolwarm',alpha=0.5,  errorbar=None)
           

for m in axC.patches:
    axC.annotate(f'{m.get_height():2f}', (m.get_x() + m.get_width()/2, m.get_height())
                 , ha='center', va='bottom', fontsize=10)
    
plt.title('Average customer ratings in each product category')
plt.xlabel=('Product_Category')
plt.ylabel=('Customer_Rating')
plt.show()
```

# **Part Three: Insight and Analysis**

### **1. Which age group has the highest average spending?**
 * ##### Answer: The 30-40 age group has the highest spending.

```python
bins = [18, 30, 40, 50, 60, 100]  
labels = ["18-30", "30-40", "40-50", "50-60", "60+"]

df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

age_group_spending = df.groupby("Age_Group")["Total_Spent"].mean().reset_index()

plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

ax = sns.barplot(x="Age_Group", y="Total_Spent", data=age_group_spending, 
                 palette="coolwarm", alpha=0.7, ci=None)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12)

plt.title("Average shopping cost in different age groups")
plt.show()
```

### **2.Is there a significant difference in shopping behavior between men and women?**
*  ##### Answer:  No, using the box plot, we can see that the average costs are almost the same in the male and female groups.


### **3. What relationship is observed between the number of purchases and the total purchase cost?**
* ##### Answer: There is a direct relationship between these variables


### **4. Which product category has the highest customer ratings?**

```python
Numeric_variables = df.drop(columns=['Customer_ID', 'Gender', 'Region', 'Product_Category', 'Age_Group'])
 
correlation_matrix = Numeric_variables.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidth=.5)

plt.title('Heatmap of Numerical Features Correlation')
plt.show()
```

```python
#Using quartiles, we divide customers into three groups:

q1 = df["Total_Spent"].quantile(0.33) 
q2 = df["Total_Spent"].quantile(0.66)  

def spending_category(value):
    if value <= q1:
        return "low_cost"
    elif value <= q2:
        return "meduim"
    else:
        return "Costly"

df["Spending_Category"] = df["Total_Spent"].apply(spending_category)


print(df["Spending_Category"].value_counts())
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))

#countplot
sns.countplot(data=df, x='Spending_Category', hue='Product_Category',
              palette='coolwarm', alpha=0.7, ax=axes[0,0])
axes[0,0].set_title("Average number of each category of customers by Product_Category")


#countplot
sns.countplot(data=df, x='Spending_Category', hue='Region',
              palette='coolwarm', alpha=0.7, ax=axes[0,1])
axes[0,1].set_title("Average number of each category of customers by region")


#boxplot
sns.boxplot(x="Spending_Category", y="Age", data=df, hue="Gender",
            palette="coolwarm", ax=axes[0,2])
axes[0,2].set_title("Box plot to show the age distribution in each cost category")

#barplot
axS = sns.barplot(x="Spending_Category", y="Purchase_Frequency", data=df, estimator=lambda y: y.mean(),
            errorbar=None, palette="coolwarm", alpha=0.7, ax=axes[1,0])
for k in axS.patches:
    axS.annotate(f'{k.get_height():.2f}', (k.get_x() + k.get_width()/2, k.get_height()),
                 ha='center', va='bottom')
axes[1,0].set_title("Bar chart for average number of purchases in each spending category")


#barplot
axU = sns.barplot(x="Spending_Category", y="Customer_Rating", data=df, estimator=lambda y: y.mean(),
                  palette="magma", alpha=0.7, errorbar=None, ax=axes[1,1])
for h in axU.patches:
    axU.annotate(f'{h.get_height():.1f}', (h.get_x() + h.get_width()/2, h.get_height()),
                 ha='center', va='bottom')
axes[1,1].set_title("Bar chart for average customer ratings in each spending category")



#piechart
spending_counts = df["Spending_Category"].value_counts()
axes[1,2].pie(spending_counts, labels=spending_counts.index, autopct="%1.1f%%", 
              colors=["lightblue", "lightgreen", "salmon"], startangle=140, 
              wedgeprops={"edgecolor": "black"})
axes[1,2].set_title("Percentage of customers by spending category")



# تنظیم فاصله بین نمودارها
plt.tight_layout()
plt.show()
```
